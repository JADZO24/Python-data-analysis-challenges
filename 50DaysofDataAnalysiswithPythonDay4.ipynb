```python
import numpy as np
rng = np.random.default_rng(seed = 42)
arr1 = rng.integers(low=10, high=21, size=(2, 3))
arr2 = rng.integers(low=10, high=21, size=(1, 3))

print("Shape of arr1:", arr1.shape)
print("Shape of arr2:", arr2.shape)
```

```python
import matplotlib.pyplot as plt
rng = np.random.default_rng(seed = 42)
arr = rng.integers(low=10, high=21, size=(100))

# Ploting the data
plt.figure(figsize=(12,8))
plt.hist(arr, bins=10)
plt.title("Hist Plot", fontsize = 30)
plt.show()
```

```python
np.mean(arr)
```

```python
np.median(arr)
```

```python
from statistics import mode

# To calculate the mode
mode = mode(arr)
print("Mode : ", mode)
```

```python
import matplotlib.pyplot as plt

# Generating data
rng = np.random.default_rng(seed = 42)
arr = rng.random((100, 3))

# Create the figure and 3D Axes object
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(projection='3d')

# Plot the data using the scatter() method
ax.scatter(arr[:,0],
           arr[:,1],
           arr[:,2])

plt.title("3D Scatter Plot", fontsize = 20)
plt.show()
```
